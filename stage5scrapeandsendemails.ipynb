```python
import pandas as pd
import copy

import selenium as sl
from selenium import webdriver

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
driverpath=ChromeDriverManager().install()

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
```

```python
df = pd.read_excel('./FinalList.xlsx',)

df
```

```python
d = webdriver.Chrome(driverpath)

n=0
for charlink in df['hyperlink']:
    profillink = charlink+'/profile'
    peoplelink = charlink+'/people'
    
    d.get(profillink)

    #get emails
    try:
        WebDriverWait(d,10).until(EC.presence_of_element_located((By.XPATH,"//*[contains(text(),'Email: ')]")))
        email1 = d.find_element(By.XPATH,"//*[contains(text(),'Email: ')]/..//dd").text

    except:
        email1 = None        
        print('Email1 failed')
    
    try:
        WebDriverWait(d,10).until(EC.presence_of_element_located((By.XPATH,"//*[contains(text(),'Address For Service email:')]")))
        email2 = d.find_element(By.XPATH,"//*[contains(text(),'Address For Service email:')]/..//dd").text
        if email1==email2:
            email2 = None
    except:
        email2 = None        
        print('Email2 failed')

    #get name
    try:
        WebDriverWait(d,10).until(EC.presence_of_element_located((By.TAG_NAME,"h1")))
        charname = d.find_element(By.TAG_NAME,"h1").text.capitalize()

    except:
        charname = None        
        print('Charname failed')

    #change page
    d.get(peoplelink)

    #get people
    try:
        WebDriverWait(d,10).until(EC.presence_of_element_located((By.CLASS_NAME,"card-body")))
        people = ""
        peoplecards = d.find_elements(By.CLASS_NAME,"card-body")
        i = len(peoplecards)
        l = copy.deepcopy(i)
        for person in peoplecards:
            fullname=person.text[:person.text.find('\n')].capitalize()
            firsname=person.text[:person.text.find(' ')].capitalize()
            if (i != 1) and i != l:
                people=people+', ' +firsname
            elif i == l:
                people = firsname
            else:
                people=people+' and '+firsname
            i-=1

    except:
        people = 'there'        
        print('people failed')
       
    ## APPEND DATA ##
    df.loc[n,['Name','email','service email','Names']] = [charname, email1,email2,people]

    n+=1

d.quit()

df
```

```python
df.to_excel('FinalList.xlsx')
```
